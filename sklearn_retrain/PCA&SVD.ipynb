```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
```

```python
X = iris.data
y = iris.target
```

```python
pca = PCA(n_components=2)
pca = pca.fit(X)
X_dr = pca.transform(X)
X_dr.shape
```

```python
# 可视化
plt.figure()
plt.scatter(X_dr[y==0,0],X_dr[y==0,1],c="red",label=iris.target_names[0])
plt.scatter(X_dr[y==1,0],X_dr[y==1,1],c="green",label=iris.target_names[1])
plt.scatter(X_dr[y==2,0],X_dr[y==2,1],c="orange",label=iris.target_names[2])
plt.title("PCA of IRIS dataset")
plt.legend()
plt.show()
```

```python
# 循环实现
color = ["red","black","orange"]
plt.figure()
for i in range(3):
    plt.scatter(X_dr[y==i,0],X_dr[y==i,1],c=color[i],label=iris.target_names[i])
plt.title("PCA of IRIS dataset")
plt.legend()
plt.show()
```

```python
from sklearn.datasets import fetch_lfw_people
from sklearn.decomposition import PCA
import numpy as np
import pandas as pd
```

```python
faces = fetch_lfw_people(min_faces_per_person=60)
```

```python
X = faces.data
```

```python
pca = PCA(150).fit(X)
X_dr = pca.transform(X)
X_dr.shape
```

```python
X_inverse = pca.inverse_transform(X_dr)
```

```python
X_inverse.shape
```

```python
fig,axes = plt.subplots(4,8,figsize=(10,6),subplot_kw={"xticks":[],"yticks":[]})
for i,ax in enumerate(axes.flat):
    ax.imshow(X_inverse[i,:].reshape(62,47),cmap="gray")
```

```python
fig,axes = plt.subplots(2,10,figsize=(10,2.5),subplot_kw={"xticks":[],"yticks":[]})
for i in range(10):
    axes[0,i].imshow(faces.images[i,:,:],cmap="binary_r")
    axes[1,i].imshow(X_inverse[i].reshape(62,47),cmap="binary_r")
```

```python
fig,axes = plt.subplots(2,10,figsize=(10,3),subplot_kw={"xticks":[],"yticks":[]})
for i in range(10):
    axes[0,i].imshow(faces.images[i,:,:],cmap="binary_r")
    axes[1,i].imshow(X_inverse[i,:].reshape(62,47),cmap="binary_r")
```

```python
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier as RFC
from sklearn.model_selection import cross_val_score
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv(r"D:\titanic\digit_recognizor.csv")
```

```python
data.head()
```

```python
data.shape
```

```python
y = data.iloc[:,0]
X = data.iloc[:,1:]
```

```python
X.shape
```

```python
y.shape
```

```python
pca_line = PCA().fit(X)
plt.figure(figsize=[20,5])
plt.plot(np.cumsum(pca_line.explained_variance_ratio_))
plt.xlabel("number of components after dimension reduction")
plt.ylabel("cumulative explained variance")
plt.legend()
plt.show()
```

```python
score = []
for i in range(1,101,10):
    X_dr = PCA(i).fit_transform(X)
    once = cross_val_score(RFC(n_estimators=10,random_state=0),X_dr,y,cv=10).mean()
    score.append(once)
    
plt.figure(figsize=[20,5])
plt.plot(range(1,101,10),score)
plt.show()
```

```python
score = []
for i in range(10,25,1):
    X_dr = PCA(i).fit_transform(X)
    once = cross_val_score(RFC(n_estimators=10,random_state=0),X_dr,y,cv=10).mean()
    score.append(once)
    
plt.figure(figsize=[20,5])
plt.plot(range(1,101,10),score)
plt.show()
```

```python
X_dr = PCA(21).fit_transform(X)
cross_val_score(RFC(n_estimators=10,random_state=0),X_dr,y,cv=5).mean()
```

```python
from sklearn.neighbors import KNeighborsClassifier as KNN
cross_val_score(KNN(),X_dr,y,cv=5).mean()
```

```python
score = []
for i in range(10):
    X_dr = PCA(21).fit_transform(X)
    once = cross_val_score(KNN(i+i),X_dr,y,cv=5).mean()
    score.append(once)
    
plt.figure(figsize=[20,5])
plt.plot(range(10),score)
plt.show()
```

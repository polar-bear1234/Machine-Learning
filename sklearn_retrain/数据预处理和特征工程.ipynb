```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn.model_selection import cross_val_score
```

```python
X_fschi = SelectKBest(chi2,k=300).fit_transform(X_fsvar,y)
```

```python
X_fschi.shape
```

```python
cross_val_score(RandomForestClassifier(n_estimators=10,random_state=0),X_fschi,y,cv=5).mean()
```

```python
chivalue,pvalue_chi = chi2(X_fsvar,y)
```

```python
chivalue.shape[0]
```

```python
(pvalue_chi > 0.5).sum()
```

```python
k = chivalue.shape[0] - (pvalue_chi > 0.5).sum()
```

```python
X_fschi = SelectKBest(chi2,k=(chivalue.shape[0]-(pvalue_chi > 0.5).sum())).fit_transform(X_fsvar,y)
cross_val_score(RandomForestClassifier(n_estimators=10,random_state=0),X_fschi,y,cv=10).mean()
```

```python
from sklearn.feature_selection import f_classif
```

```python
F,pvalues_f = f_classif(X_fsvar,y)
```

```python
k = F.shape[0] - (pvalues_f > 0.05).sum()
```

```python
k
```

```python
X_fsF = SelectKBest(f_classif,k=(F.shape[0]-(pvalues_f > 0.05).sum())).fit_transform(X_fsvar,y)
cross_val_score(RandomForestClassifier(n_estimators=10,random_state=0),X_fsF,y,cv=5).mean()
```

```python
from sklearn.feature_selection import mutual_info_classif as MIC
```

```python
result = MIC(X_fsvar,y)
(result>0).sum()
```

```python
k = result.shape[0] - (result <= 0).sum()
```

```python
k
```

```python
X_fsF = SelectKBest(MIC,k = (result.shape[0]-(result <= 0).sum())).fit_transform(X_fsvar,y)
cross_val_score(RandomForestClassifier(n_estimators=10,random_state=0),X_fsF,y,cv=10).mean()
```

```python
from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import RandomForestClassifier as RFC
```

```python
RFC_ = RFC(n_estimators=10,random_state=0)
```

```python
X_embedded = SelectFromModel(RFC_,threshold=0.005).fit_transform(X,y)
```

```python
X_embedded.shape
```

```python
import matplotlib.pyplot as plt
import numpy as np
threshold = np.linspace(0,(RFC_.fit(X,y).feature_importance_).max(),20)
score_e = []
for i in threshold:
    X_embedded = SelectFromModel(RFC_,threshold=i).fit_transform(X,y)
    score_e = cross_val_score(RFC_,X_embedded,y,cv=10).mean()
    
plt.figure()
plt.plot(threshold,score_e)
plt.show()
# 嵌入法选出来的特征比方差过滤要少，时间要更短，交叉验证的结果比过滤后的结果要好
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
```

```python
rng = np.random.RandomState(1)  # éšæœºç§å­
x = np.sort(5*rng.rand(80,1),axis=0)
```

```python
y = np.sin(x).ravel()
```

```python
y.shape
```

```python
y[::5] += 3*(0.5-rng.rand(16))
```

```python
plt.figure()
plt.scatter(x,y,s=20,edgecolor="black",c="darkorange",label="data")
plt.legend()
plt.show()
```

```python
regr_1 = DecisionTreeRegressor(max_depth=2)
regr_2 = DecisionTreeRegressor(max_depth=5)
regr_1.fit(x,y)
regr_2.fit(x,y)
```

```python
x_test = np.arange(0.0,5.0,0.01)[:,np.newaxis]
```

```python
y_1 = regr_1.predict(x_test)
```

```python
y_2 = regr_2.predict(x_test)
```

```python
plt.figure()
plt.scatter(x,y,s=20,edgecolor="black",c="darkorange",label="data")
plt.plot(x_test,y_1,color="cornflowerblue",label="max_depth=2",linewidth=2)
plt.plot(x_test,y_2,color="yellowgreen",label="max_depth=5",linewidth=2)
plt.xlabel("data")
plt.ylabel("target")
plt.title("Decision Tree Regression")
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
import numpy as np
```

```python
data = pd.read_csv("D:/titanic/data.csv")
```

```python
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
data.drop(["Name","Cabin","Ticket"],inplace=True,axis=1)
```

```python
data["Age"] = data["Age"].fillna(data["Age"].mean())
```

```python
data = data.dropna(axis=0)
```

```python
data.info()
```

```python
data.head()
```

```python
labels = data["Embarked"].unique().tolist()
```

```python
labels
```

```python
data["Embarked"] = data["Embarked"].apply(lambda x:labels.index(x))
```

```python
data.loc[:,"Sex"] = (data.loc[:,"Sex"] == "male").astype("int")
# astypeèƒ½å¤Ÿå°†ä¸€ä¸ªpandaså¯¹è±¡è½¬æ¢ä¸ºæŸç§ç±»åž‹ï¼Œå’Œapplyï¼ˆintï¼ˆxï¼‰ï¼‰ä¸åŒï¼Œastypeå¯ä»¥å°†æ–‡æœ¬ç±»åž‹è½¬æ¢ä¸ºæ•°å­—ï¼Œ
# ç”¨è¿™ä¸ªæ–¹å¼å¯ä»¥å¾ˆæ–¹ä¾¿çš„å°†äºŒåˆ†ç±»ç‰¹å¾è½¬æ¢ä¸º0å’Œ1
```

```python
X = data.loc[:,data.columns != "Survived"]
y = data.loc[:,data.columns == "Survived"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
Xtrain,Xtest,Ytrain,Ytest = train_test_split(X,y,test_size=0.3,random_state=30)
```

```python
Xtrain.shape
```

```python
for i in [Xtrain,Xtest,Ytrain,Ytest]:
    i.index = range(i.shape[0])
```

```python
Xtrain.head()
```

```python
clf = DecisionTreeRegressor(random_state=25)
clf = clf.fit(Xtrain,Ytrain)
score = clf.score(Xtest,Ytest)
score
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(clf,Xtrain,Ytrain,cv=10).mean()
```

```python
# åšä¸€ä¸ªæ ‘æ·±å­¦ä¹ æ›²çº¿
tr = []
te = []
for i in range(10):
    clf = DecisionTreeRegressor(max_depth=i+1,random_state=25)
    clf = clf.fit(Xtrain,Ytrain)
    tr.append(clf.score(Xtest,Ytest))
    te.append(cross_val_score(clf,X,y,cv=10).mean())
print(max(te))
print(max(tr))
plt.figure()
plt.plot(range(1,11),tr,color="red",label="fit_train")
plt.plot(range(1,11),te,color="green",label="cross_val_score")
plt.ylabel("R^2")
plt.legend()
plt.show()
```

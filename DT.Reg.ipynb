实例：泰坦尼克号幸存者的预测

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier  # 只能处理数字，不能是文字
import matplotlib.pyplot as plt
from sklearn.model_selection import GridSearchCV
```

```python
data = pd.read_csv("D:/titanic/data.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
# 筛选特征
data.drop(["Cabin","Name","Ticket"],inplace=True,axis=1) # 删除完毕的表覆盖原表，默认是False
# data = data.drop(["Cabin","Name","Ticket"],inplace=False)
```

```python
data.tail()
```

```python
# 处理缺失值
data["Age"] = data["Age"].fillna(data["Age"].mean())
```

```python
data.info()
```

```python
data = data.dropna() # 默认axis=0---行      （全局）
```

```python
data.info()
```

```python
data.head()
```

```python
labels = data["Embarked"].unique().tolist()
```

```python
labels
```

```python
data["Embarked"] = data["Embarked"].apply(lambda x: labels.index(x))
```

```python
#labels.index("S")
```

```python
#(data["Sex"] =="male").astype("int")
```

```python
data.loc[:,"Sex"] = (data["Sex"] == "male").astype("int")
```

```python
data.head()
```

```python
x = data.iloc[:,data.columns != "Survived"]  # 划分训练集
```

```python
y = data.iloc[:,data.columns == "Survived"]  # 划分测试集
```

```python
from sklearn.model_selection import train_test_split
Xtrain,Xtest,Ytrain,Ytest = train_test_split(x,y,test_size=0.3)
```

```python
Xtrain.index = range(Xtrain.shape[0]) # 纠正索引
```

```python
# 统一写
for i in [Xtrain,Xtest,Ytrain,Ytest]:
    i.index = range(i.shape[0])
```

```python
# Xtrain
```

开始跑模型

```python
clf = DecisionTreeClassifier(random_state=25)  # 随机数任何数都可以
```

```python
clf = clf.fit(Xtrain,Ytrain)
score = clf.score(Xtest,Ytest)
score
```

使用交叉验证

```python
from sklearn.model_selection import cross_val_score
clf = DecisionTreeClassifier(random_state=25)
score = cross_val_score(clf,x,y,cv=10).mean()
score
```

```python
# 做一个树深学习曲线
import matplotlib.pyplot as plt
tr = []
te = []
for i in range(10):
    clf = DecisionTreeClassifier(random_state=25
                                ,max_depth=i+1
                                ,criterion="entropy"
                                )
    clf = clf.fit(Xtrain,Ytrain)
    score_tr = clf.score(Xtrain,Ytrain)
    score_te = cross_val_score(clf,x,y,cv=10).mean()
    tr.append(score_tr)
    te.append(score_te)
print(max(te))
plt.plot(range(1,11),tr,color="red",label="train")
plt.plot(range(1,11),te,color="blue",label="test")
plt.xticks(range(1,11))  # 修改横坐标的标尺
plt.legend()
plt.show()
```

```python
population_dict = {"BeiJing":2154,
                   "ShangHai":2424,
                   "ShenZhen":1303,
                   "HangZhou":981}
GDP_dict = {"BeiJing":30320,
            "ShangHai":32680,
            "ShenZhen":24222,
            "HangZhou":13468}
population = pd.Series(population_dict)
GDP = pd.Series(GDP_dict)
city_info = pd.DataFrame({"population":population,"GDP":GDP})
city_info["Per_GDP"] = city_info["GDP"]/city_info["population"]
city_info
```

```python
# 根据人均GDP对表格排序
city_info.sort_values(by="Per_GDP") # 升序
```

```python
city_info.sort_values(by="Per_GDP",ascending=False) # 降序
```

```python
def make_df(cols,ind):
    data = {c:[str(c)+str(i) for i in ind] for c in cols}
    # print(data)
    return pd.DataFrame(data,ind)
make_df("ABC",range(3))
```

```python
# @ ----例子
import seaborn as sns
planets = sns.load_dataset("planets")
planets.shape
```

```python
planets.head()
```

```python
planets.describe()
```

```python
decade = 10*(planets["year"]//10)
decade = decade.astype(str)+"s"
decade.name = "decade"
decade.head()
```

```python
planets.groupby(["method",decade]).sum()
```

```python
planets.groupby(["method",decade])[["number"]].sum().unstack().fillna(0)
```

```python
# @例子-----泰坦尼克号
import seaborn as sns
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
titanic.describe()
```

```python
titanic.groupby("sex")[["survived"]].mean()
```

```python
titanic.groupby(["sex","class"])["survived"].aggregate("mean").unstack()
```

```python
titanic.pivot_table("survived",index="sex",columns="class")  # 数据透视表pivot_table
```

```python
titanic.pivot_table("survived",index="sex",columns="class",aggfunc="mean",margins=True)  # 数据透视表
```

```python
titanic.pivot_table(index="sex",columns="class",aggfunc={"survived":"sum","fare":"mean"})  # 数据透视表
```
